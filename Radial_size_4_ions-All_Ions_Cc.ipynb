```python
#Graphs as a function of  position of beam 

beam_rpositions=np.array([401.5, 402, 402.5,403.5,404])
rate=[]
for i in range (4):
    
    rates_xscan=np.array([rates_xscan_401s[i],rates_xscan_402[i],rates_xscan_402s[i],rates_xscan_403s[i],rates_xscan_404[i]])
    rate.append(rates_xscan) 
    
    fig, ax = plt.subplots(1, figsize = (5, 5))
    ax.scatter(beam_rpositions, rates_xscan)
    ax.set_xlabel("Beam Radial Position")
    ax.set_ylabel("Transition Rate")
    ax.set_title(f'Ion {i+1}')

   
```

```python
sigma=[]
fig, ax = plt.subplots(1, figsize = (5,5))
ax.scatter(beam_rpositions, rate[0], label = 'Ion 1', color = 'm')
ax.scatter(beam_rpositions, rate[1], label = 'Ion 2', color = 'b')
ax.scatter(beam_rpositions, rate[2], label = 'Ion 3', color = 'y')
ax.scatter(beam_rpositions, rate[3], label = 'Ion 4', color = 'g')

popt, pcov = curve_fit(func.Gaussian, beam_rpositions, rate[0], p0 = [402.5, 0.5, 25, 0], bounds = ([-np.inf, -np.inf, -np.inf, -.1], [np.inf, np.inf, np.inf, .1]))
x = np.arange(398, 408, .01)
y = func.Gaussian(x, *popt)
plt.plot(x, y, 'm')
print(popt[1])
sigma.append(popt[1])

popt, pcov = curve_fit(func.Gaussian, beam_rpositions, rate[1], p0 = [402.5, 0.5, 25, 0], bounds = ([-np.inf, -np.inf, -np.inf, -.1], [np.inf, np.inf, np.inf, .1]))
x = np.arange(398, 408, .01)
y = func.Gaussian(x, *popt)
plt.plot(x, y, 'b')
print(popt[1])
sigma.append(popt[1])

popt, pcov = curve_fit(func.Gaussian, beam_rpositions, rate[2], p0 = [402.5, 0.5, 25, 0], bounds = ([-np.inf, -np.inf, -np.inf, -.1], [np.inf, np.inf, np.inf, .1]))
x = np.arange(398, 408, .01)
y = func.Gaussian(x, *popt)
plt.plot(x, y, 'y')
print(popt[1])
sigma.append(popt[1])

popt, pcov = curve_fit(func.Gaussian, beam_rpositions, rate[3], p0 = [402.5, 0.5, 25, 0], bounds = ([-np.inf, -np.inf, -np.inf, -.1], [np.inf, np.inf, np.inf, .1]))
x = np.arange(398, 408, .01)
y = func.Gaussian(x, *popt)
plt.plot(x, y, 'g')
sigma.append(popt[1])

print(popt[1])


plt.legend()
plt.xlabel("beam radial positions")
plt.ylabel("transition rates")
```
